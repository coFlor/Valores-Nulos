```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Realizo el llamado a los datos que se encuentran en el csv

```python
data= pd.read_csv('Ventas_totales.csv')
data
```

*DATA1*
data 1 lo ocupé para el skiprows pero debido a que en las cabeceras está la información necesaria, esta ocasion no hubo necesidad de hacer desaparecer a una de las filas

```python
data1= pd.read_csv('Ventas_totales.csv', skiprows=0)
data1
```

*DATA2*
Para este lo que hice fue mandar a llamar a data1 y con ello a su contenido, para hacer desaparecer a las columnas llamadas dentro del 
.drop

```python
data2 = data1.drop({'otros', 'otros_medios', 'bebidas', 'panaderia', 'lacteos', 'carnes', 'verduleria_fruteria'}, axis=1)
data2
```

Con la muestra de valores_nulos siguiente, se mostrará los datos nulos que aun se encuentran en la información

```python
valores_nulos=data2.isnull().sum()
valores_nulos
```

Para llenar la columna de "subtotal_ventas_alimentos_bebidas" realicé el mandado a llenar aquellas celdas sin información con la media del resto de celdas que sí tienen contenido.

```python
data3=data2.copy()
```

```python
data3["subtotal_ventas_alimentos_bebidas"]=data3["subtotal_ventas_alimentos_bebidas"].fillna(round(data2["subtotal_ventas_alimentos_bebidas"].mean(),1))
```

```python
valores_nulos=data3.isnull().sum()
valores_nulos
```

La nueva columna a cubrir es "tarjetas_credito" y esta la decidí llenar con .fillna("SIN EXISTENCIA") ya que al ser un dato numerico y unico, no podría rellenarlo asi como asi, estaria dando falsos datos.

```python
#segundo método de sustitucion de valores nulos
#sustituir valores nulos con mediana
data3["tarjetas_credito"] = data3["tarjetas_credito"].fillna("SIN EXISTENCIA")
```

```python
valores_nulos=data3.isnull().sum()
valores_nulos
```

Para las siguientes columnas de "salon_ventas" y "alimentos_preparados_rotiseria" usé el llenado con valores, puse 0 para representar mejor la inexistencia de los mismos

```python
data3["salon_ventas"]=data3["salon_ventas"].fillna(000)
data3["alimentos_preparados_rotiseria"]=data3["alimentos_preparados_rotiseria"].fillna(000)
```

```python
valores_nulos=data3.isnull().sum()
valores_nulos
```

En el caso de la información de "tarjetas_debito" lo que hice fue rellenarlo con NULL como muestra de que es nulo, no existe dicha información

```python
data3["tarjetas_debito"] = data3["tarjetas_debito"].fillna("NULL")
```

```python
valores_nulos=data3.isnull().sum()
valores_nulos
```

Apartir de aquí hice uso de ffill y bfill para terminar con "almacen" y "electronicos_articulos_hogar"

```python
data3["almacen"] = data3["almacen"].fillna(method="ffill")
```

```python
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
data3["electronicos_articulos_hogar"]=data3["electronicos_articulos_hogar"].fillna(method="bfill")
```

```python
valores_nulos=data3.isnull().sum()
valores_nulos
```

Para los ultimos nulos correspondientes a "indumentaria_calzado_textiles_hogar" hice uso de .dropna(), aunque esto causa perdida de datos extra, hace que ya no haya existencia de nulos.

```python
data3=data2.dropna()
data3
```

```python
valores_nulos=data3.isnull().sum()
valores_nulos
```

Hago el llamado final de .to_csv para mostrar en ello que ya no existen valores nulos

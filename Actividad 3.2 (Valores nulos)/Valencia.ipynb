Uso skiprows para quitar la fila de las variables en el idioma inglés para tenerlas de acuerdo al diccionario que tenemos.

```python
data1= pd.read_csv('Valencia.csv', skiprows=1)
data1
```

```python
valores_nulos=data1.isnull().sum()
valores_nulos
```

Las siguientes columnas decidí ponerlas con un .fillna para string ya que considero que no puede usarle otro tipo de filtro de relleno para nulos ya que además de dar información erronea, se considera dar enunciados necesarios.

Se puede notar la diferencia en cada una.
En los datos 4,5 y 6 he considerado a "N/A" ya que el archivo muestra datos así entonces para no salirme de la línea que llevan, los iguale para que se tuviese nula descripcion.

El resto, la mayoría tienes .fillna distintos con el afán de no ser repetitivo y que a al ves sea exclusivo.
En baños y dormitorios he considerado "Sin Info" más que nada para mostrar de otra forma la nula información que hay.

```python
data2= data1["descripción"] = data1["descripción"].fillna("Sin descripción")
```

```python
data3= data1["descripción general del vecindario"] = data1["descripción general del vecindario"].fillna("No cuenta con una")
```

```python
data4=data1["tiempo_respuesta_host"] = data1["tiempo_respuesta_host"].fillna("N/A")
```

```python
data5=data1["tasa_respuesta_host"] = data1["tasa_respuesta_host"].fillna("N/A")
```

```python
data6=data1["tasa_de_aceptación_host"] = data1["tasa_de_aceptación_host"].fillna("N/A")
```

```python
data7=data1["host_is_superhost"] = data1["host_is_superhost"].fillna("No mencionado")
```

```python
data8=data1["baños"] = data1["baños"].fillna("Sin info")
```

```python
data9=data1["dormitorios"] = data1["dormitorios"].fillna("Sin info")
```

```python

data10=data1["camas"] = data1["camas"].fillna("Sin info")
data11=data1["precio"] = data1["precio"].fillna("No asignado")
data12=data1["tiene_disponibilidad"] = data1["tiene_disponibilidad"].fillna("No hay detalles")
data13=data1["primera_revisión"] = data1["primera_revisión"].fillna("Sin fecha")
data14=data1["última_revisión"] = data1["última_revisión"].fillna("Sin fecha")
```

```python
valores_nulos=data1.isnull().sum()
valores_nulos
```

En los siguientes datos considero a .fillna(0) ya que estas en sus columnas considera información numerica, y al ser nulos, como es un "review" pues ni la media podría aplicarse ni strings para que se mantenga en orden a como está la información, por ello, para mostrar un nulo review en estos consideré "0" en representación de que no tienen calificaciones por la comunidad

```python
data15=data1["review_scores_rating"]=data1["review_scores_rating"].fillna(0)
```

```python

data16=data1["review_scores_accuracy"]=data1["review_scores_accuracy"].fillna(0)
data17=data1["review_scores_cleanliness"]=data1["review_scores_cleanliness"].fillna(0)
data18=data1["review_scores_checkin"]=data1["review_scores_checkin"].fillna(0)
data19=data1["review_scores_comunicación"]=data1["review_scores_comunicación"].fillna(0)
data20=data1["review_scores_location"]=data1["review_scores_location"].fillna(0)
data21=data1["review_scores_value"]=data1["review_scores_value"].fillna(0)
```

```python
valores_nulos=data1.isnull().sum()
valores_nulos
```

"reseñas_por_mes" Al contemplar datos individuales considero que el relleno mediante media podría ser una opción, pero no la mejor ya que estaríamos dando datos falsos a fin de cuentas, consdiero en especial en este rellenarlo con string "--"

```python
data22=data1["reseñas_por_mes"] = data1["reseñas_por_mes"].fillna("--")
```
